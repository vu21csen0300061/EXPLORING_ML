```python
print(titanic_data.head(5))
```

```python
sex=pd.get_dummies(titanic_data["Sex"],drop_first=True)
sex.head(5)
```

```python

embark=pd.get_dummies(titanic_data["Embarked"],drop_first=True)
embark.head(5)
```

```python
pcl=pd.get_dummies(titanic_data["Pclass"],drop_first=True)
pcl.head(5)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pcl],axis=1)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['PassengerId','Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop("Pclass",axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
x=titanic_data.drop("Survived",axis=1)
y=titanic_data["Survived"]
```

```python

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
reg.fit(X_train.values,Y_train.values)
```

```python
predictions=reg.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report
classification_report(predictions,Y_test.values)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(predictions,Y_test.values)
```
